```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('files/gender.csv')

data['Class'] = data['Gender'].apply(lambda x: 'r' if x == 'F' else 'b')
data.head(5)
```

```python
fig, ax = plt.subplots()

data.iloc[:25].plot.scatter(x='Height', y='Hair length', c='Class', ax=ax)
ax.axhline(20)
x = np.array([150, 190])
ax.plot(x, x - 145, c='y')
ax.plot(x, 2*x - 310, c='g')
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataset = data.iloc[:25]

X = dataset[['Height', 'Hair length']]
y = dataset['Gender']
y = np.array([0 if value == 'M' else 1 for value in y])
```

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X, y)
```

```python
X_test = np.random.rand(10000,2)
X_test = X_test*(50, 120) + (150, 0)

y_pred = clf.predict(X_test)
```

```python
fig, ax = plt.subplots()

ax.scatter(x=X_test[:,0], y=X_test[:,1], c=y_pred, alpha=.5)
y_color = ['r' if i == 1 else 'b' for i in y]
ax.scatter(x=X['Height'], y=X['Hair length'], c=y_color)
ax.set_xlabel('Height')
ax.set_ylabel('Hair length')
```
